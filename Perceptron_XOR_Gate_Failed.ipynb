```python
def activation_function(sum_input):
    sum_input_bias = sum_input
 
    # Binary Threshold Activation Function:
    if sum_input_bias >= 0.5:
        result = 1
    else:
        result = 0
        
    return result
```

```python
# A single neuron:
def perceptron(i_1,i_2,w_1,w_2,bb):
    input_1 = i_1
    input_2 = i_2
    weight_1 = w_1
    weight_2 = w_2
    bias = bb
    
    sop_with_bias = weight_1*input_1 + weight_2*input_2 + bias # Sum of 'Input*Weight+bias'
    output = activation_function(sop_with_bias) # Calling Activation Function
    return output
```

```python
weight_matrix = [3,1] # Weight initialization
b=1 # Bias initialization
lr = 0.1 # Learning Rate
```

```python
# XOR Gate::
# Input features:
feature_vector_x = [[0,0],
                    [0,1],
                    [1,0],
                    [1,1]]
feature_vector_y = [0,1,1,0] # Expected outputs 
```

```python
correct_prediction = 0
epoch_no = 0

while True:
    if correct_prediction == 4: 
        break
        
    epoch_no += 1
    print("Epoch",epoch_no,":\n-----------\n")
    
    correct_prediction = 0
    iteration_no = 0
    for feature,label in zip(feature_vector_x,feature_vector_y):
        iteration_no += 1
        print("Iteration",iteration_no,":")
        
        x1=feature[0]
        x2=feature[1]

        #print("weight_matrix =",weight_matrix)
        w1 = weight_matrix[0]
        w2 = weight_matrix[1]

        target_output = label
        pred_output = perceptron(x1,x2,w1,w2,b)
        error = target_output - pred_output

        print("Target =",target_output,"; Predicted =",pred_output,"; Error =",error,"\n")
        
        if target_output == pred_output:
            correct_prediction += 1
            continue

        if sum(weight_matrix) == 0:
            rate_w1 = rate_w2 = 0.5
        else:
            rate_w1 = w1/sum(weight_matrix)
            rate_w2 = w2/sum(weight_matrix)
            
        weight_matrix = []
            
        delta_w1 = error*lr*rate_w1
        w1 = w1 + delta_w1
        weight_matrix.append(w1)
        
        delta_w2 = error*lr*rate_w2
        w2 = w2 + delta_w2
        weight_matrix.append(w2)
        
        b = b + error*lr
        
        #print("w1 =",w1,", w2=",w2,", b=",b)
        #print("weight_matrix =",weight_matrix)
        
    accuarcy = (correct_prediction/len(feature_vector_y))*100
    print("\nAccuracy =",accuarcy,"%")
    print("   ------------------------   \n")
    #break
```
